```python
# to convert this string '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# into list ['Action','Adventure','Fantasy','Science Fiction'], we will make a function
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# apply this function to genres
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch)
```

```python
movies.head()
```

```python
# ṭo convert string in overview into a list 
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# to concatenate the lists......before that we have to solve some problems
# STEP 3: remove inbetween spaces in each value inside every list so that similar names do not create confusion for the model.

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
# create a new column TAGS as the concatenated column of all 4 columns 
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags'] =new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df.head()
```

```python
# to convert entire tag into lower case
new_df['tags'] =new_df['tags'].apply(lambda x :x.lower())
```

```python
new_df.head()
```

```python
# vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
# to remove similar words
import nltk
from nltk.stem.porter import PorterStemmer  
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] =new_df['tags'].apply(stem)
```

```python
new_df
```

```python
#  calculate cos distance not eucladian distance.for high dimensional distance.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
# here similarity is an arrary of similarity values from each movie to every movies including itself 
# present in the data ie.4806x4806 therefor this array is square and have diagonal elements as 1.
```

```python
similarity
```

```python
# make a fun. which returns the 5 closest movies to a given movie.
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list: 
        print(new_df.iloc[i[0]].title)
```
